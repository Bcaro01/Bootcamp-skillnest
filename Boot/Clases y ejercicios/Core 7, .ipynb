### Interpretación de datos:

1. type: hace referencia al tipo de propiedad al cual esta enfocado, como parcelas, casas, casas pariadas, departamentos, etc
2. text: es la descripción de la propiedad
3. year_build: es el año en el cual se construyó el inmueble
4. beds: cantidad de habitaciones
5. baths: cantidad total de baños en la propiedad (baths_full + baths_half)
6. baths_full: cantidad de baños con duchas que incluye la propiedad
7. baths_half: cantidad de baños sin duchas que incluye la propiedad
8. garage: capacidad de autos que caben en el garage
9. lot_sqft: area total de la propiedad
10. sqft : hace referencia al total construido en el terreno
11. stories: hace referencia a la cantidad de pisos
12. lastSoldPrice: último precio al cual fue vendido la propiedad
13. soldOn: fecha de la última venta de la propiedad
14. listPrice: precio de venta de la propiedad al momento de la recopilación del dataset
15. status: estado de venta de la propiedad

### Limpieza de los datos

```python
# Eliminación de duplicados
df = df.drop_duplicates()

# Eliminación filas null de 'year_built'
    # se eliminan las filas debido a que no es posible rellenar los datos con valores Extrapolados ni estadístidos, debido a que no se puede deducir gracias a otros tados.

df = df.dropna(subset=['year_built'])
df['year_built'] = df['year_built'].astype(int)

# Caso beds:
df[df['beds'].isna() == True] # arroja que todos los NaN corresponden a land, lo que tiene sentido porque no hay habitaciones en los campos abiertos.
                            # desde esto, se cambian los valores nulos por 0 y al mismo tiempo se cambia el formato.
df['beds'] = df['beds'].fillna(0).astype(int)

# Caso baths_full:
df[df['baths_full'].isna() == True] # en todos estos casos, baths_full y baths_half suman 0 al total de baths, por lo que se reemplazarán por 0 y su tipo al mismo tiempo
df['baths'] = df['baths'].astype(int)
df['baths_full'] = df['baths_full'].fillna(0).astype(int)
df['baths_half'] = df['baths_half'].fillna(0).astype(int)
#df.info()

# Caso garage:
df[df['garage'].isna() == True] 
df.loc[(df['garage'].isna()) & (df['type'] == 'land'), 'garage'] = 0 # las tierras que no tienen construcción se pone con garage 0
df['garage'] = df['garage'].fillna(round(df['garage'].mean())).astype(int)


# Caso lot_sqft
# Rellenar según correlación entre sqft y lot_sqft
df['lot_sqft'] = df['lot_sqft'].fillna(df['sqft'] * (df['lot_sqft'].mean() / df['sqft'].mean()))
df['lot_sqft'] = df['lot_sqft'].fillna(df['lot_sqft'].median())

# Remover espacios innecesarios y convertir a minúsculas
df['type'] = df['type'].str.strip().str.lower()
df['status'] = df['status'].str.strip().str.lower()

# para casos de 'type': 'land'
f1 = (df['sqft'].isna() == True)
f2 = (df['type'] == 'land')
f3 = (df['stories'].isna() == True)
f4 = (df['garage'].isna() == True)

df.loc[(df['sqft'].isna()) & (df['type'] == 'land'), 'sqft'] = 0 # se identifica cuales tienen sqft nan y se rellenan con 0, ya que son tierras, lo mismo con stories
df.loc[(df['stories'].isna()) & (df['type'] == 'land'), 'stories'] = 0

# Caso lot_sqft
df['sqft'] = df['sqft'].fillna(df['lot_sqft'] * (df['sqft'].mean() / df['lot_sqft'].mean()))
df['sqft'] = df['sqft'].fillna(df['lot_sqft'].median())

df.info()





```

### Transformación de datos

```python
# Se calcularán los nuevos campos valor por pie cuadrado ($/sqft) y edad de al propiedad al día de hoy
df['$/sqft'] = round(df['listPrice'] / df['sqft'], 2)
df['antique_prop'] = 2025 - df['year_built']
```

### Análisis de datos

```python
type_status_group = df.groupby(['type', 'status']).agg(
    total_properties = ('type', 'count'),
    avg_value_per_sqft = ('$/sqft', 'mean'),
    antique_avg=('antique_prop', 'mean')
).reset_index().sort_values('avg_value_per_sqft', ascending = False)

print(type_status_group)
```
