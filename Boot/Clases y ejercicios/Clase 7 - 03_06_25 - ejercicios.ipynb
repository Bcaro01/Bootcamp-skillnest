# Caso:

Eres parte de una empresa que se encarga a alquilar propiedades en diferentes ciudades. Se te pide hacer un analisis sobre la data de las propiedades. Debes responder a las siguientes preguntas:

1. Cuantas propiedades registradas tiene la empresa.
2. Cual es el promedio de renta de las propiedades (market rent).
3. Cual es el promedio de renta segun tipologia (tipology).
4. Cual es el promedio de renta por ciudad.
5. Cual es el promedio de renta por tipologia y ciudad.
6. Muestra una comparacion (promedio) de m2 y habitaciones por pais.
7. Muestra una comparacion (promedio y desviacion) de m2 y habitaciones por pais.
8. Muestra una comparacion (promedio y desviacion) de m2 y habitaciones de los distritos de Lima.
9. Muestra una comparacion de las rentas de las ciudades de Perú y Colombia.
10. Que pais y ciudad en mas importante para la empresa.

```python
import pandas as pd
```

```python
path = 'C:/Users/Bcaro/OneDrive/Documentos/VSC/Skillnest/Boot/Clases y ejercicios/Clase 7 - Ejercicios.xlsx'
df = pd.read_excel(path)
df
```

```python
# pregunta 1
print(f"Hay un total de {df.groupby(['prop_id'])['prop_id'].count().sum()} propiedades.")
```

```python
# pregunta 2
print(f"La renta media de las propiedades es de {df['Market Rent'].mean()} unidades monetarias.")
```

```python
#pregunta 3
a = df.groupby(['Typology'])['Market Rent'].mean().sort_values(ascending=False)
print("La renta promedio de las propiedades según tipo es:\n", a)
```

```python
#pregunta 4
b = df.groupby(['City'])['Market Rent'].mean().sort_values(ascending=False)
print("La renta promedio de las propiedades según tipo es:\n", b)
```

```python

#pregunta 5
c = df.groupby(['City', 'Typology'])['Market Rent'].mean().round(2)
print("La renta promedio de las propiedades según ciudad y tipo es:\n", c.unstack())
```

```python
#pregunta 6
d = df.groupby(['Country']).agg({'m2': ['mean'], 'Beds': ['mean']}).round(2)
print("La comparativa de m2 y de cantidad de camas por país es:\n", d)
```

```python
#pregunta 7
e = df.groupby(['Country']).agg({'m2': ['mean', 'std'], 'Beds': ['mean', 'std']}).round(2)
print("La comparativa de m2 y de cantidad de camas por país es:\n", e)
```

```python
# Pregunta 8
f1 = df['City'] == 'Lima'
df_f = df[f1]
f = df_f.groupby(['District']).agg({'m2': ['mean', 'std'], 'Beds': ['mean', 'std']}).round(2)
print("La comparativa de m2 y de cantidad de camas por distrito en Lima es:\n", f)
```

```python
#pregunta 9
f2 = (df['Country'] == 'Peru') | (df['Country'] == 'Colombia')

df_f2 = df[f2]
g = df_f2.groupby(['Country', 'City']).agg({'Market Rent': ['mean']}).round(2)
print("La comparativa de m2 y de cantidad de habitaciones por país y tipo es:\n", g)
```

```python
#pregunta 10
df['Profit'] = df['Market Rent'] - df['Utilities']
df.groupby(['Country', 'City']).agg({'Profit': ['sum', 'mean', 'std']}).round(2).sort_values(by=('Profit', 'sum'), ascending=False).head(5)
```
