### Filtrado de datos

```python
# se filtra para obtener las ventas mayores a 50
# Se asume que las "Ventas" hacen referencia al dato "Total Amount", ya que tambien se podr√≠a asumir que se refiere al hecho de la venta o a la cantidad de unidades vendidas.

df[df['Total Amount'] > 50]
```

```python
# se filtra para obtener las precio menor a 0.5
# Se asume que las "Precio" hacen referencia al dato "Price per Unit"
# Como no existe ningun valor menor que 0.5, se filtra por montos menores a 30 unidaes monetarias

df[df['Price per Unit'] < 30]
```

```python
# Debido a que en el enunciado se solicita filtar por productos de tipo "Manzana" con ventas mayores a 30 UM (Unidades Monetarias), y esto no existe,
# Se filtrar√° por productos de tipo "Clothing" en donde sus ventas sean mayores	30. con el mismo supuesto de que las ventas se refiere al Total Amount.

filtered_df = df.query('(`Product Category` == "Clothing") & (`Total Amount` >= 30)')
filtered_df
```
